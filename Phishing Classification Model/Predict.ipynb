```
import sys
import os
import json
import pandas as pd
import numpy as np
import optparse
from keras.callbacks import TensorBoard
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.preprocessing.text import Tokenizer
from collections import OrderedDict
from pprint import pprint
import tensorflow as tf
import pickle


# Carpeta de drive on es guarda el Model (mou la carpeta Phishing Classification al teu drive)
# Si vols mantenir la estructura hauras de crear primer la carpeta "Colab Notebooks" en el teu drive
# i despres copiar la carpeta "Phishing Classification" al "Colab Notebooks"
model_path = "/content/drive/MyDrive/Colab Notebooks/Phishing Classification Model/model"
# Cargar model amb tensorflow
pretrained_model = tf.keras.models.load_model(model_path)
# Fitxer de drive on es guarda el tokenizer (no cal tocar-lo si tens la mateixa estructura)
tokenizer_path = "/content/drive/MyDrive/Colab Notebooks/Phishing Classification Model/tokenizer.pickle"
# Cargar tokenizer
with open(tokenizer_path, 'rb') as handle:
    tokenizer = pickle.load(handle)
```

```
# L'URL que vols comprovar
input_url = "https://www.upc.edu/ca"
# Codificar input amb tokenizer
url = tokenizer.texts_to_sequences([input_url])
max_log_length = 2083
url = sequence.pad_sequences(url, maxlen=max_log_length)
# Obtenir la probabilitat
pretrained_model.predict(url)[0]
# El resultat indica la probabilitat de que el url donat es un web de phishing
```
